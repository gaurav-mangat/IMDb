```python
# Q30. Identify any outliers in the movie runtime data. How would you handle them
```

```python
Q1 = df['Runtime (Minutes)'].quantile(0.25)
Q3 = df['Runtime (Minutes)'].quantile(0.75)
IQR = Q3 - Q1
threshold = 1.5
outliers = df[(df['Runtime (Minutes)'] < Q1 - threshold * IQR) | (df['Runtime (Minutes)'] > Q3 + threshold * IQR)]["Runtime (Minutes)"]
print(outliers)
```
